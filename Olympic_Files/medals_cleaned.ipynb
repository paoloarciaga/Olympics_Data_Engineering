```python
# Import dependencies
import pandas as pd
import numpy as np
from pathlib import Path
import re
pd.set_option('max_colwidth', 400)
```

```python
# File to Load (Remember to Change These)
Olympic_games_medals_data = Path("Olympic_Files/olympic_medals.csv")
Olympic_games_olympic_results_data = Path("Olympic_Files/olympic_results.csv")

# Read medals and results Data File and store into Pandas DataFrames
medals_data = pd.read_csv(Olympic_games_medals_data)
medals_data.head()
```

```python
medals_data.info()
```

```python
medals_data[['first_name', 'last_name']] = medals_data ["athlete_full_name"].str.split(' ', n=1, expand=True)
medals_data['last_name'] = medals_data['last_name'].str.lower()
medals_data['first_name'] = medals_data['first_name'].str.capitalize()
medals_data['last_name'] = medals_data['last_name'].str.capitalize()
medals_data.head(100)
```

```python
# Replace missing values with an empty string
medals_data['athlete_full_name'] = medals_data['athlete_full_name'].fillna('')

# Splitting last name into components
medals_data[['first_name', 'last_name']] = medals_data["athlete_full_name"].apply(lambda x: pd.Series(str(x).split(' ', 1)))

# Capitalizing each component of the last name
medals_data['last_name'] = medals_data['last_name'].apply(lambda x: ' '.join([part.capitalize() for part in str(x).split()]))

# Capitalizing first name
medals_data['first_name'] = medals_data['first_name'].str.capitalize()

medals_data.head()
```

```python
# Function to split and capitalize last name
def split_and_capitalize_last_name(full_name):
    if isinstance(full_name, str):
        # Splitting the full name into parts
        name_parts = full_name.split(' ')
        if len(name_parts) > 1:
            # Captializing the last name parts
            last_name_parts = [part.capitalize() for part in name_parts[1:]]  # Consider all parts after the first
            return ' '.join(last_name_parts)
        elif len(name_parts) > 1:
            # Captializing the last name parts
            last_name_parts = [part.capitalize() for part in name_parts[1:]]  # Consider all parts after the first
            return '-'.join(last_name_parts)
        elif len(name_parts) == 1:
            # Capitalize single-word last name
            return name_parts[0].capitalize()  
    return ''

# Apply the function to extract last names
medals_data['last_name'] = medals_data['athlete_full_name'].apply(split_and_capitalize_last_name)

medals_data.head()
```

```python
# Drop rows where 'athlete_full_name' column is an empty string
medals_data = medals_data[medals_data['athlete_full_name'] != '']

# Now you can apply the function to extract last names
medals_data['last_name'] = medals_data['athlete_full_name'].apply(split_and_capitalize_last_name)


medals_data.head()
```

```python
# Drop unwanted columns  
medals_data = medals_data.drop(['athlete_full_name', 'country_code', 'participant_title'], axis=1)

medals_data.head()
```
