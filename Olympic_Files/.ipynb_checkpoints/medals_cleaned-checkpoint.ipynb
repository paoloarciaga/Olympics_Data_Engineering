```python
# Import dependencies
import pandas as pd
import numpy as np
from pathlib import Path
import re
pd.set_option('max_colwidth', 400)
```

```python
# File to Load (Remember to Change These)
Olympic_games_medals_data = Path("Olympic_Files/olympic_medals.csv")
Olympic_games_olympic_results_data = Path("Olympic_Files/olympic_results.csv")

# Read medals and results Data File and store into Pandas DataFrames
medals_data = pd.read_csv(Olympic_games_medals_data)
medals_data.head()
```

```python
medals_data.info()
```

```python
medals_data[['first_name', 'last_name']] = medals_data ["athlete_full_name"].str.split(' ', n=1, expand=True)
medals_data['last_name'] = medals_data['last_name'].str.lower()
medals_data['first_name'] = medals_data['first_name'].str.capitalize()
medals_data['last_name'] = medals_data['last_name'].str.capitalize()
medals_data.head(100)
```

```python
# Replace missing values with an empty string
medals_data['athlete_full_name'] = medals_data['athlete_full_name'].fillna('')

# Splitting last name into components
medals_data[['first_name', 'last_name']] = medals_data["athlete_full_name"].apply(lambda x: pd.Series(str(x).split(' ', 1)))

# Capitalizing each component of the last name
medals_data['last_name'] = medals_data['last_name'].apply(lambda x: ' '.join([part.capitalize() for part in str(x).split()]))

# Capitalizing first name
medals_data['first_name'] = medals_data['first_name'].str.capitalize()

medals_data.head()
```

```python
# Function to split and capitalize last name
def split_and_capitalize_last_name(full_name):
    if isinstance(full_name, str):
        # Splitting the full name into parts
        name_parts = full_name.split(' ')
        if len(name_parts) > 1:
            # Captializing the last name parts
            last_name_parts = [part.capitalize() for part in name_parts[1:]]  # Consider all parts after the first
            return ' '.join(last_name_parts)
        elif len(name_parts) > 1:
            # Captializing the last name parts
            last_name_parts = [part.capitalize() for part in name_parts[1:]]  # Consider all parts after the first
            return '-'.join(last_name_parts)
        elif len(name_parts) == 1:
            # Capitalize single-word last name
            return name_parts[0].capitalize()  
    return ''

# Apply the function to extract last names
medals_data['last_name'] = medals_data['athlete_full_name'].apply(split_and_capitalize_last_name)

medals_data.head(50)
```

```python
medals_data.columns
```

```python
# Drop rows where 'athlete_full_name' column is an empty string
medals_data = medals_data[medals_data['athlete_full_name'] != '']

# Now you can apply the function to extract last names
medals_data['last_name'] = medals_data['athlete_full_name'].apply(split_and_capitalize_last_name)


medals_data.head()
```

```python
# Drop unwanted columns  
medals_data = medals_data.drop(['athlete_full_name', 'country_code', 'participant_title'], axis=1)

medals_data.head(50)
```

```python
# Grouping the data by specified columns
grouped = medals_data.groupby(['discipline_title', 'slug_game', 'country_name'])

# Creating a list to store dictionaries of aggregated data
aggregated_data_list = []

# Iterating through each group
for group_name, group_data in grouped:
    # Extracting data for athlete 1
    athlete1_data = group_data.iloc[0]
    
    # Checking if there is only one row in the group
    if len(group_data) == 1:
        # If only one row exists, set athlete2 data to None
        athlete2_data = None
    else:
        # If more than one row exists, extract data for athlete 2
        athlete2_data = group_data.iloc[1]
        # If athlete 1 and athlete 2 have the same name, take the next athlete
        if athlete1_data['first_name'] == athlete2_data['first_name'] and athlete1_data['last_name'] == athlete2_data['last_name']:
            athlete2_data = group_data.iloc[2] if len(group_data) > 2 else None
    
    # Creating a dictionary for the aggregated data
    aggregated_data = {
        'discipline_title': group_name[0],
        'slug_game': group_name[1],
        'country_name': group_name[2],
        'country_3_letter_code': group_data['country_3_letter_code'].iloc[0],
        'athlete1_first_name': athlete1_data['first_name'] if athlete1_data['participant_type'] == 'Athlete' else None,
        'athlete1_last_name': athlete1_data['last_name'] if athlete1_data['participant_type'] == 'Athlete' else None,
        'athlete1_url': athlete1_data['athlete_url'] if athlete1_data['participant_type'] == 'Athlete' else None,
        'athlete2_first_name': athlete2_data['first_name'] if athlete2_data is not None and athlete2_data['participant_type'] == 'Athlete' else None,
        'athlete2_last_name': athlete2_data['last_name'] if athlete2_data is not None and athlete2_data['participant_type'] == 'Athlete' else None,
        'athlete2_url': athlete2_data['athlete_url'] if athlete2_data is not None and athlete2_data['participant_type'] == 'Athlete' else None,
        'participant_type': athlete1_data['participant_type'] if athlete1_data['participant_type'] == 'Athlete' else None,
        'discipline_title': athlete1_data['discipline_title'],
        'event_title': athlete1_data['event_title'],
        'event_gender': athlete1_data['event_gender'],
        'medal_type': athlete1_data['medal_type']
    }
    
    # Appending the dictionary to the list
    aggregated_data_list.append(aggregated_data)

# Creating a DataFrame from the list of aggregated data
aggregated_data = pd.DataFrame(aggregated_data_list)

# Define custom sorting order
custom_sort_order = [
    'beijing-2022', 'tokyo-2020', 'pyeongchang-2018', 'rio-2016', 'sochi-2014',
    'london-2012', 'vancouver-2010', 'beijing-2008', 'turin-2006', 'athens-2004',
    'salt-lake-city-2002', 'sydney-2000', 'nagano-1998', 'atlanta-1996',
    'lillehammer-1994', 'barcelona-1992', 'albertville-1992', 'seoul-1988',
    'calgary-1988', 'los-angeles-1984', 'sarajevo-1984', 'moscow-1980',
    'lake-placid-1980', 'montreal-1976', 'innsbruck-1976', 'munich-1972',
    'sapporo-1972', 'mexico-city-1968', 'grenoble-1968', 'tokyo-1964',
    'innsbruck-1964', 'rome-1960', 'squaw-valley-1960', 'melbourne-1956',
    'cortina-d-ampezzo-1956', 'helsinki-1952', 'oslo-1952', 'london-1948',
    'st-moritz-1948', 'berlin-1936', 'garmisch-partenkirchen-1936',
    'los-angeles-1932', 'lake-placid-1932', 'amsterdam-1928', 'st-moritz-1928',
    'paris-1924', 'chamonix-1924', 'antwerp-1920', 'stockholm-1912',
    'london-1908', 'st-louis-1904', 'paris-1900', 'athens-1896'
]

# Apply custom sorting order to slug_game column
aggregated_data['slug_game'] = pd.Categorical(aggregated_data['slug_game'], categories=custom_sort_order, ordered=True)

# Sort the DataFrame by slug_game column
aggregated_data_sorted = aggregated_data.sort_values(by='slug_game')

# Displaying the sorted
aggregated_data_sorted.head()
```

```python
aggregated_data_sorted.to_csv('medals_cleaned.csv', index=False)
```
