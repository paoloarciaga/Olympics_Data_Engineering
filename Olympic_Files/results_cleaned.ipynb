```python
# Replace missing values with an empty string
results_data['athlete_full_name'] = results_data['athlete_full_name'].fillna('')

# Splitting last name into components
results_data[['first_name', 'last_name']] = results_data["athlete_full_name"].apply(lambda x: pd.Series(str(x).split(' ', 1)))

# Capitalizing each component of the last name
results_data['last_name'] = results_data['last_name'].apply(lambda x: ' '.join([part.capitalize() for part in str(x).split()]))

# Capitalizing first name
results_data['first_name'] = results_data['first_name'].str.capitalize()

results_data.head()
```

```python
def split_and_capitalize_last_name(full_name):
    if isinstance(full_name, str):
        # Splitting the full name into parts
        name_parts = full_name.split(' ')
        # Captializing the last name parts
        if len(name_parts) > 1:
            last_name_parts = []
            for part in name_parts[1:]:
                if '-' in part:
                    hyphen_parts = part.split('-')
                    capitalized_parts = [p.capitalize() for p in hyphen_parts]
                    last_name_parts.extend(capitalized_parts)
                else:
                    last_name_parts.append(part.capitalize())
            return ' '.join(last_name_parts)

# Test the function with a sample value
print(split_and_capitalize_last_name("Cooper Woods-topalovic"))
print(split_and_capitalize_last_name("Almida De val"))

```

```python
# Fill missing values with empty string
results_data['athletes'].fillna('', inplace=True)

# Initialize lists to store extracted data
athlete_data = []

# Iterate over each row in the DataFrame
for idx, entry in enumerate(results_data['athletes']):
    # Split the string by '), (' to separate each athlete's information
    athlete_info = entry.strip('[]').split("), ('")
    # Extract athlete names and URLs from each entry
    for info in athlete_info:
        # Split each entry by ', ' to separate athlete name and URL
        name_url = info.split("', '")
        # Check if name_url has at least two elements
        if len(name_url) >= 2:
            # Extract athlete name and URL
            athlete_name = name_url[0].strip("('")
            athlete_url = name_url[1].strip("')")
            # Append the extracted data to the list
            athlete_data.append({'First Name': athlete_name.split()[0],
                                  'Last Name': split_and_capitalize_last_name(athlete_name),
                                  'URL': athlete_url,
                                  'participant_type': results_data.iloc[idx]['participant_type']})

# Create a DataFrame from the extracted data
df = pd.DataFrame(athlete_data)

df.head()
```

```python
import re

# Assuming results_data is a Pandas DataFrame
athlete_data_list = []

# Iterate over each string representation of list of tuples
for athlete_list_str in results_data['athletes']:
    # Handle empty strings
    if not athlete_list_str.strip():
        continue
    
    # Split the string into tuples
    athlete_list = athlete_list_str.split("), (")
    
    # Remove leading and trailing brackets for the first and last tuples
    if len(athlete_list) > 0:
        athlete_list[0] = athlete_list[0].lstrip("[(")
    if len(athlete_list) > 1:
        athlete_list[-1] = athlete_list[-1].rstrip(")]")
    
    # Iterate over each tuple in the list
    for athlete_data_str in athlete_list:
        # Try to match patterns where name and URL are separated by different delimiters
        match = re.match(r"['\"]([^'\"]*)['\"](?:, (['\"][^'\"]*['\"]))?", athlete_data_str)
        if match:
            athlete_name = match.group(1)
            athlete_url = match.group(2)
            
            # Split the name into first and last name
            name_parts = athlete_name.split(' ')
            if len(name_parts) >= 2:
                first_name = name_parts[0]
                last_name = ' '.join(name_parts[1:])
            else:
                # If name doesn't have both first and last name, use full name as first name
                first_name = athlete_name
                last_name = ""
            
            # If URL is missing, set it to None
            if athlete_url:
                athlete_url = athlete_url.strip('"')
            else:
                athlete_url = None
            
            athlete_data_list.append({
                'First Name': first_name,
                'Last Name': last_name,
                'URL': athlete_url
            })
        else:
            print(f"Failed to extract name and URL from: {athlete_data_str}")

# Convert the list of dictionaries into a DataFrame
athlete_df = pd.DataFrame(athlete_data_list)


athlete_df.head()
```

```python
import ast

# Function to extract athlete information
def separate_athletes(row):
    try:
        athletes_info = row['athletes']
        if athletes_info:
            athlete_info = ast.literal_eval(athletes_info)
            if isinstance(athlete_info, list) and len(athlete_info) >= 2:
                athlete1 = athlete_info[0]
                athlete2 = athlete_info[1]
                athlete1_first_name, athlete1_last_name = athlete1[0].split(maxsplit=1)
                athlete2_first_name, athlete2_last_name = athlete2[0].split(maxsplit=1)
                return pd.Series([
                    athlete1_first_name, athlete1_last_name, athlete1[1], 
                    athlete2_first_name, athlete2_last_name, athlete2[1]
                ])
            else:
                return pd.Series([None, None, None, None, None, None])
        else:
            return pd.Series([None, None, None, None, None, None])
    except Exception as e:
        print(f"Error processing row: {row}")
        print(f"Error message: {e}")
        return pd.Series([None, None, None, None, None, None])

# Apply the function to each row
results_data[['athlete1_first_name', 'athlete1_last_name', 'athlete1_url', 
              'athlete2_first_name', 'athlete2_last_name', 'athlete2_url']] = results_data.apply(separate_athletes, axis=1)



results_data.head()
```

```python
def format_last_name(last_name):
    if last_name is None:
        return None
    # Split the last name by hyphen or space if present
    parts = [part.capitalize() for part in last_name.split('-' if '-' in last_name else ' ')]
    # Join the parts to form the formatted last name
    return '-'.join(parts) if '-' in last_name else ' '.join(parts)

# Apply the formatting function to 'athlete1_last_name' and 'athlete2_last_name' columns
results_data['athlete1_last_name'] = results_data['athlete1_last_name'].apply(format_last_name)
results_data['athlete2_last_name'] = results_data['athlete2_last_name'].apply(format_last_name)

results_data.head()
```
