```python
# Import dependencies
import pandas as pd
import numpy as np
from pathlib import Path
pd.set_option('max_colwidth', 400)
```

```python
Olympic_games_olympic_results_data = Path("Olympic_Files/olympic_results.csv")

# Read medals and results Data File and store into Pandas DataFrames
results_data = pd.read_csv(Olympic_games_olympic_results_data)
results_data.head(50)
```

```python
results_data.columns
```

```python
# Count NaN values in the 'athlete_full_name' column
nan_count = results_data['athlete_full_name'].isna().sum()

print("Number of NaN values in 'athlete_full_name' column:", nan_count)
```

```python
# Replace missing values with an empty string
results_data['athlete_full_name'] = results_data['athlete_full_name'].fillna('')

# Splitting last name into components
results_data[['first_name', 'last_name']] = results_data["athlete_full_name"].apply(lambda x: pd.Series(str(x).split(' ', 1)))

# Capitalizing each component of the last name
results_data['last_name'] = results_data['last_name'].apply(lambda x: ' '.join([part.capitalize() for part in str(x).split()]))

# Capitalizing first name
results_data['first_name'] = results_data['first_name'].str.capitalize()

results_data.head(50)
```

```python
def split_and_capitalize_last_name(full_name):
    if isinstance(full_name, str):
        # Splitting the full name into parts
        name_parts = full_name.split(' ')
        # Captializing the last name parts
        if len(name_parts) > 1:
            last_name_parts = []
            for part in name_parts[1:]:
                if '-' in part:
                    hyphen_parts = part.split('-')
                    capitalized_parts = [p.capitalize() for p in hyphen_parts]
                    last_name_parts.extend(capitalized_parts)
                else:
                    last_name_parts.append(part.capitalize())
            return ' '.join(last_name_parts)
        elif len(name_parts) == 1 and name_parts[0]:  # Check if name_parts[0] is not empty
            # Special case for last names like "De val"
            if name_parts[0].lower() == 'de':  # Check if the first part is 'De' in any case
                return name_parts[0].capitalize() + ' ' + name_parts[1].capitalize() if len(name_parts) > 1 else name_parts[0].capitalize()
            else:
                return name_parts[0].capitalize()  # Capitalize single-word last name
    return ''

# Test the function with a sample value
print(split_and_capitalize_last_name("Cooper Woods-topalovic"))
print(split_and_capitalize_last_name("Almida De val"))

results_data.head()
```

```python
# Fill missing values with empty string
results_data['athletes'].fillna('', inplace=True)

# Initialize lists to store extracted data
athlete_data = []

# Iterate over each row in the DataFrame
for idx, entry in enumerate(results_data['athletes']):
    # Split the string by '), (' to separate each athlete's information
    athlete_info = entry.strip('[]').split("), ('")
    # Extract athlete names and URLs from each entry
    for info in athlete_info:
        # Split each entry by ', ' to separate athlete name and URL
        name_url = info.split("', '")
        # Check if name_url has at least two elements
        if len(name_url) >= 2:
            # Extract athlete name and URL
            athlete_name = name_url[0].strip("('")
            athlete_url = name_url[1].strip("')")
            # Append the extracted data to the list
            athlete_data.append({'First Name': athlete_name.split()[0],
                                  'Last Name': split_and_capitalize_last_name(athlete_name),
                                  'URL': athlete_url,
                                  'participant_type': results_data.iloc[idx]['participant_type']})

# Create a DataFrame from the extracted data
df = pd.DataFrame(athlete_data)

df.head()
```

```python
# Merge the dataframes
merged_df = pd.merge(results_data, df, on='participant_type')
merged_df.head()
```
